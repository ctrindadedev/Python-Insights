```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
#NÃºmero de linhas e Colunas
df.shape
```

```python
#Verificando os dados do DataSet
df.info()
```

```python
#2)Estudar a distribuiÃ§Ã£o das variÃ¡veis numÃ©ricas, calculando, para cada coluna, as principais estatÃ­sticas descritivas de posiÃ§Ã£o (mÃ©dia, mediana, quartis, etc.) e de dispersÃ£o (std, IQR, etc.).
df_desc=df.describe()
df_desc
```

```python
#Defininindo uma nova colunar para analisarmos o IQR juntamente aos demais dados utilizando a funÃ§Ã£o LOC
#iqr Q3 - Q1
df_desc.loc["IQR"] = df_desc.loc["75%"] - df_desc.loc["25%"]
df_desc
```

```python
#Verificando OUTLIERS (Dados fora da curva com grÃ¡fico)
for col in df:
    sns.histplot(data=df, x=col, kde=True).set_title(f"DistribuiÃ§Ã£o da variÃ¡vel {col}")
    plt.show()
```

```python
#ContinuaÃ§Ã£o da visualizaÃ§Ã£o de outliers (com IQR), agora identificando quais sÃ£o seus indices
for col in df.drop(columns='quality'): #Dropado pois estamos analisando os dados para obsever a influÃªncia na qualidade.
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR = Q3-Q1
    aux_outliers =df[(df[col] < Q1-(IQR*1.5)) | (df[col]>Q3+(IQR*1.5))]
    indices_outliers = aux_outliers.index.tolist()
    if len(indices_outliers) >= 1:
            print(f"A coluna {col} tem {len(indices_outliers)} outliers!")
            print("\nOs indices deles sÃ£o: \n") 
            print(indices_outliers)

    else:
            print(f"Acoluna {col} nÃ£o tem outliers!")   
    print()
    print("-"*80)
    print()
```
